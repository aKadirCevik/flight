```python
#df = pd.concat(file_dict)
df.reset_index(drop=True, inplace=True)
df["label"] = 1
df.sort_values("time")
df_turkey = df
```

```python
# lon: boylam 26 - 45
# lat: enlemv 36 - 42
```

```python
df_turkey = df[(df.lon < 100) & (df.lon > 10) & (df.lat < 100) & (df.lat > 10)]
df_turkey = df_turkey.dropna()
df_turkey
```

# Unix timestamp'in saat ve tarihe çevrilmesi

```python
time_type = []
date_type = []

for i in range(0, len(df_turkey)):
    t_type = time.strftime("%X", time.localtime(df_turkey["time"].iloc[i]))
    d_type = time.strftime("%D", time.localtime(df_turkey["time"].iloc[i]))
    time_type.append(t_type)
    date_type.append(d_type)

df_turkey["time_t"] = time_type
df_turkey["date"] = date_type
    
df_turkey
```

```python
feauture_list = ["icao24", "date", "time_t", "time", "lat", "lon", "heading", "velocity", "baroaltitude", "geoaltitude", "label"]
df_full_data =  df_turkey[feauture_list]
df_full_data
```

```python
uniqueValues = (df_full_data['icao24']).unique()
len(uniqueValues)
for i in range(0, len(uniqueValues)):
    for j in range(0, len(df_full_data)):
        if( df_full_data['icao24'].iloc [j] == uniqueValues[i]):
            df_full_data['icao24'].iloc [j] = i + 1
            
            
df_full_data
```

```python
df_full_data =  df_full_data[["icao24", "time", "lat", "lon", "heading", "velocity", "baroaltitude", "geoaltitude", "label"]]
df_full_data
```

```python
df_full_data.to_csv('df_turkey_full.csv', index=False)
```

# Callsign'e göre bir günde kaç farklı uçağın geçtiğinin hesaplanması

```python
uniqueValues = (df_turkey['icao24']).unique()
len(uniqueValues)
# uniqueValues
```

```python
column_names = ["icao24", "date", "time", "lat", "lon", "heading", "velocity", "baroaltitude", "geoaltitude", "class"]
df_deneme = pd.DataFrame(columns = column_names)

for i in range(0, len(uniqueValues)):
    df_turkey_uni = df_turkey[df_turkey["icao24"] == uniqueValues[i]]
    
    
    # lat, lon ve time kolonlarını bir hücrede birleştirme 
    # (vektör olarak ifade etme)
    time_arr = df_turkey_uni["time_t"].to_list()
    lat_arr = df_turkey_uni["lat"].to_numpy()
    lon_arr = df_turkey_uni["lon"].to_numpy()
    head_arr = df_turkey_uni["heading"].to_numpy()
    vel_arr = df_turkey_uni["velocity"].to_numpy()
  #  squawk_arr = df_turkey_uni["squawk"].to_numpy()
    baro_arr = df_turkey_uni["baroaltitude"].to_numpy()
    geo_arr = df_turkey_uni["geoaltitude"].to_numpy()
    
    df_deneme.at[i, 'icao24'] = uniqueValues[i]
    df_deneme.at[i, 'date'] = df_turkey_uni["date"].iloc[0]
    df_deneme.at[i, 'lat'] = lat_arr
    df_deneme.at[i, 'lon'] = lon_arr
    df_deneme.at[i, 'time'] = time_arr
    df_deneme.at[i, 'heading'] = head_arr
    df_deneme.at[i, 'velocity'] = vel_arr
   # df_deneme.at[i, 'squawk'] = squawk_arr
    df_deneme.at[i, 'baroaltitude'] = baro_arr
    df_deneme.at[i, 'geoaltitude'] = geo_arr
    df_deneme.at[i, 'class'] = 0
    
    
df_deneme    
```

```python
df_resut = df_deneme[["time", "lat", "lon", "heading", "velocity", "baroaltitude", "geoaltitude", "class"]]
df_resut
```

```python
df_resut.to_csv('df_turkey.csv', index=False)
```
