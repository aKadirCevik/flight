# Callsign'e göre bir günde kaç farklı uçağın geçtiğinin hesaplanması

```python
uniqueValues = (df_turkey_attack['icao24']).unique()
#len(uniqueValues)
# uniqueValues
```

```python
column_names = ["icao24", "date", "time", "lat", "lon", "heading", "velocity", "baroaltitude", "geoaltitude", "class"]
df_deneme = pd.DataFrame(columns = column_names)

for i in range(0, len(uniqueValues)):
    df_turkey_uni = df_turkey_attack[df_turkey_attack["icao24"] == uniqueValues[i]]
    
    
    # lat, lon ve time kolonlarını bir hücrede birleştirme 
    # (vektör olarak ifade etme)
    time_arr = df_turkey_uni["time_t"].to_list()
    lat_arr = df_turkey_uni["lat"].to_numpy()
    lon_arr = df_turkey_uni["lon"].to_numpy()
    head_arr = df_turkey_uni["heading"].to_numpy()
    vel_arr = df_turkey_uni["velocity"].to_numpy()
  #  squawk_arr = df_turkey_uni["squawk"].to_numpy()
    baro_arr = df_turkey_uni["baroaltitude"].to_numpy()
    geo_arr = df_turkey_uni["geoaltitude"].to_numpy()
    
    df_deneme.at[i, 'icao24'] = uniqueValues[i]
    df_deneme.at[i, 'date'] = df_turkey_uni["date"].iloc[0]
    df_deneme.at[i, 'lat'] = lat_arr
    df_deneme.at[i, 'lon'] = lon_arr
    df_deneme.at[i, 'time'] = time_arr
    df_deneme.at[i, 'heading'] = head_arr
    df_deneme.at[i, 'velocity'] = vel_arr
   # df_deneme.at[i, 'squawk'] = squawk_arr
    df_deneme.at[i, 'baroaltitude'] = baro_arr
    df_deneme.at[i, 'geoaltitude'] = geo_arr
    df_deneme.at[i, 'class'] = 0
    
    
# df_deneme    
```

```python
df_resut = df_deneme[["time", "lat", "lon", "heading", "velocity", "baroaltitude", "geoaltitude", "class"]]
# df_resut
```

```python
df_resut.to_csv('df_turkey_attack.csv', index=False)
```
