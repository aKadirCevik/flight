```python
# use glob to get all the csv files
# in the folder
path = "C:\\Users\\ncevik\\Desktop\\TİK2\\Flight\\Data\\"
csv_files = glob.glob(os.path.join(path, "*.csv"))

# print(csv_files)
```

```python
file_dict = {}
counter = 1
for file in csv_files:
    df = pd.read_csv(file)
    file_dict[counter] = df
    counter += 1
```

```python
df = pd.concat(file_dict)
df.reset_index(drop=True, inplace=True)
df["label"] = 1
df.sort_values("time")
```

```python
uniqueValues = (df['icao24']).unique()
```

```python
for i in range(0, len(uniqueValues)):
    for j in range(0, len(df)):
        if( df['icao24'].iloc [j] == uniqueValues[i]):
            df['icao24'].iloc [j] = i + 1 
            
df
```
