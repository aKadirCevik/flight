```python
import pandas as pd
import numpy as np
from numpy import array
import os
import sys
```

```python
sys.path.append(os.getcwd()+'\\test_data\\resultDF_lastFilter_deneme.csv')  # system path'ine adres eklenir.   
path = sys.path[-1]  #en son eklenen adres kullanılır.
dataset = pd.read_csv(path)
dataset
```

```python
counter = 0
for i in range(0, len(dataset)):
    if isinstance(dataset["primary"].iloc[i], str): 
    # if type(dataset["MAIL"].iloc[i]) == str:
        counter += 1
        temp = dataset["primary"].iloc[i]
        dataset["primary"].iloc[i] = counter
    for j in range(i+1, len(dataset)):
        if temp == dataset["primary"].iloc[j]:
            dataset["primary"].iloc[j] = counter

dataset_class_name = dataset.groupby("primary")
mail_list = dataset_class_name.groups.keys()
# print(mail_list)
```
