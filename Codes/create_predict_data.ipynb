```python
import pandas as pd
from datetime import datetime
import numpy as np
import time

df = pd.read_csv('C:\\Users\\ncevik\\Desktop\\TİK2\\Flight\\Data\\states_2021-12-06-20.csv') # Use Tab to separate
```

```python
df_turkey = df[(df.lon < 45) & (df.lon > 26) & (df.lat < 42) & (df.lat > 36)]
df_turkey = df_turkey.dropna()
df_turkey
```

```python
time_type = []
date_type = []

for i in range(0, len(df_turkey)):
    t_type = time.strftime("%X", time.localtime(df_turkey["time"].iloc[i]))
    d_type = time.strftime("%D", time.localtime(df_turkey["time"].iloc[i]))
    time_type.append(t_type)
    date_type.append(d_type)

df_turkey["time_t"] = time_type
df_turkey["date"] = date_type
df_turkey["label"] = 1
    
df_turkey
```

```python
df_full_data =  df_turkey[["icao24", "date", "time_t", "time", "lat", "lon", "heading", "velocity", "baroaltitude", "geoaltitude", "label"]]
df_full_data
```

```python
uniqueValues = (df_turkey['icao24']).unique()
#len(uniqueValues)
for i in range(0, len(uniqueValues)):
    for j in range(0, len(df_full_data)):
        if( df_full_data['icao24'].iloc [j] == uniqueValues[i]):
            df_full_data['icao24'].iloc [j] = i + 1
            
            
df_full_data
```

```python
df_full_data =  df_full_data[["icao24", "time", "lat", "lon", "heading", "velocity", "baroaltitude", "geoaltitude", "label"]]
df_full_data
```
